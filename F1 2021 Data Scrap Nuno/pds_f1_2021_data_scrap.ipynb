# F1 2021 Data Sets

```python
#Libraries
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request
```

### 2021 race results

```python
source = urllib.request.urlopen('https://www.formula1.com/en/results.html/2021/races.html').read()
soup = BeautifulSoup(source,'lxml')
table = soup.find_all('table')[0] 
df = pd.read_html(str(table), flavor='bs4', header=[0])[0]
df.drop(["Unnamed: 0","Unnamed: 7"],axis=1, inplace=True)
df.head(50)
```

### 2021 Driver Standings

```python
source = urllib.request.urlopen('https://www.formula1.com/en/results.html/2021/drivers.html').read()
soup = BeautifulSoup(source,'lxml')
table = soup.find_all('table')[0] 
df = pd.read_html(str(table), flavor='bs4', header=[0])[0]
df.drop(["Unnamed: 0","Unnamed: 6"],axis=1, inplace=True)
df.head()
```
