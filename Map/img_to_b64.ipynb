```python
import os
import base64
import pandas as pd
from PIL import Image
from io import BytesIO
```

```python
def cvt_2_base64(file_name):
    with open(file_name , "rb") as image_file :
        data = base64.b64encode(image_file.read())
    return data.decode('utf-8')
```

```python
def get_b64(directory):
    pics = []
    names = []

    for file in os.listdir(directory):
        pic = os.path.join(directory, file)
        with open(pic , "rb") as image_file:
            pic_encoded = base64.b64encode(image_file.read())
            pics.append(pic_encoded.decode('utf-8'))
            file.split('.')[0]
            names.append(file)
    
    df_b64 = pd.DataFrame({'name': names, 'b64': pics})
    
    return df_b64
```
