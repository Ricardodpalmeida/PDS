```python
def getcoords(df):
    
    all_lat = []
    all_lng = []
    
    for i, name in enumerate(df):
        coord = wiki.WikipediaPage(title=name).coordinates
        lat = float(coord[0])
        lng = float(coord[1])
        all_lat.append(lat)
        all_lng.append(lng)

    return all_lat, all_lng
```
